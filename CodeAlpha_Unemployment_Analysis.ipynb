convert date column

```python
df1[' Date']=pd.to_datetime(df1[' Date'],dayfirst=True)
df2[' Date']=pd.to_datetime(df2[' Date'],dayfirst=True)
```

Rename the column

```python
df1.rename(columns={' Estimated Unemployment Rate (%)':'Unemployment_rate'}, inplace=True)
df2.rename(columns={ ' Estimated Unemployment Rate (%)':"Unemployment_rate"}, inplace=True)
```

VISUALIZATION (DATASET(DF1))

```python
df1_avg = df1.groupby(' Date')['Unemployment_rate'].mean()

plt.figure(figsize=(10,5))
plt.plot(df1_avg.index, df1_avg.values)

plt.title("Average Unemployment Trend (Before COVID)")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate")

plt.show()
```

```python
df1.groupby('Region')['Unemployment_rate'].mean().plot(kind='bar')
plt.title("Average Unemployment by Region")
plt.show()
```

DATASET ANALYSIS DF2

```python
df2_avg = df2.groupby(' Date')['Unemployment_rate'].mean()

plt.figure(figsize=(10,5))
plt.plot(df2_avg.index, df2_avg.values)

plt.title("Unemployment Trend During COVID-19")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate")

plt.show()
```

```python
df2.groupby('Region')['Unemployment_rate'].mean().sort_values(ascending=False).head().plot(kind='bar')
plt.title("Top Regions with High Unemployment")
plt.show()
```

```python
df2['Month'] = df2[ ' Date'].dt.month

sns.boxplot(x='Month', y='Unemployment_rate', data=df2)
plt.title("Monthly Unemployment Pattern")
plt.show()
```
